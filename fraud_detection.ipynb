removing outliers

```python
import pandas as pd
# Define a function to remove outliers based on IQR
def remove_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Filter out the outliers
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Select numerical columns
numerical_columns = ["Amount", "Time"]  # Adjust based on your dataset
# Remove outliers
data_no_outliers = remove_outliers_iqr(data.copy(), numerical_columns)

print(f"Original data size: {data.shape}")
print(f"Data size after removing outliers: {data_no_outliers.shape}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix

# Assuming X_balanced and y_balanced are already available from the previous SMOTE-Tomek step

# 1. Use half of the balanced data
X_half = X_balanced[:len(X_balanced)//2]
y_half = y_balanced[:len(y_balanced)//2]

# 2. Split the half data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_half, y_half, test_size=0.2, random_state=42)

# 3. Define individual models
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
lr = LogisticRegression(random_state=42, max_iter=1000)
svm = SVC(probability=True, random_state=42)
nb = GaussianNB()

# 4. Create the soft voting ensemble classifier
ensemble_model = VotingClassifier(
    estimators=[('dt', dt), ('rf', rf), ('lr', lr), ('svm', svm), ('nb', nb)],
    voting='soft'
)

# 5. Train the ensemble model
ensemble_model.fit(X_train, y_train)

# 6. Predict on the test data
y_pred = ensemble_model.predict(X_test)

# 7. Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.feature_selection import RFE

# ... (previous imports and SMOTE-Tomek code) ...

# Assuming X_balanced and y_balanced are already available

# 1. Use half of the balanced data
X_half = X_balanced[:len(X_balanced)//2]
y_half = y_balanced[:len(y_balanced)//2]

# 2. Split the half data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_half, y_half, test_size=0.2, random_state=42)

# 3. Feature Selection using RFE
estimator = RandomForestClassifier(random_state=42)  # Use Random Forest for feature ranking
selector = RFE(estimator, n_features_to_select=10, step=1)  # Select top 10 features
selector = selector.fit(X_train, y_train)

# Apply feature selection to training and testing data
X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)

# 4. Define individual models (using selected features)
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
lr = LogisticRegression(random_state=42, max_iter=1000)
svm = SVC(random_state=42)
nb = GaussianNB()

# 5. Create the hard voting ensemble classifier
ensemble_model = VotingClassifier(
    estimators=[('dt', dt), ('rf', rf), ('lr', lr), ('svm', svm), ('nb', nb)],
    voting='hard'
)

# 6. Train the ensemble model (using selected features)
ensemble_model.fit(X_train_selected, y_train)

# 7. Predict on the test data (using selected features)
y_pred = ensemble_model.predict(X_test_selected)

# 8. Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Get the selected feature names (if needed)
selected_feature_names = X_half.columns[selector.support_]
print("\nSelected Feature Names:", selected_feature_names)
```
